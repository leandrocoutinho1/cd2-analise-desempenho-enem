Foi criada a variável `MEDIA_GERAL`, calculada como a média aritmética das cinco notas do ENEM (Ciências da Natureza, Ciências Humanas, Linguagens, Matemática e Redação).

Essa variável representa um indicador sintético de desempenho acadêmico global e será utilizada tanto na modelagem quanto nas análises comparativas envolvendo características institucionais das escolas.

```python
# Removendo registros com média igual a zero
df_valid = df_valid[df_valid["MEDIA_GERAL"] > 0].copy()
df_valid_escola = df_valid_escola[df_valid_escola["MEDIA_GERAL"] > 0].copy()

print("Shape após remoção (df_valid):", df_valid.shape)
print("Shape após remoção (df_valid_escola):", df_valid_escola.shape)
```

### 📌 Tratamento de Valores Extremos

Foram identificados 5 registros com média geral igual a zero, representando uma fração estatisticamente irrelevante da base.

Considerando que esses valores configuram extremos que podem distorcer algoritmos baseados em distância, optou-se por removê-los da análise.

```python
# Mapeamento ordinal da renda (A = menor renda, Q = maior renda)
mapa_renda = {letra: i for i, letra in enumerate(
    list("ABCDEFGHIJKLMNOPQ"), start=1)
}

df_valid["RENDA_ORDINAL"] = df_valid["Q006"].map(mapa_renda)
df_valid_escola["RENDA_ORDINAL"] = df_valid_escola["Q006"].map(mapa_renda)

df_valid["RENDA_ORDINAL"].describe()
```

```python
from sklearn.preprocessing import StandardScaler

# Padronização das variáveis de desempenho para evitar influência de escala
# O StandardScaler transforma os dados para média 0 e desvio padrão 1,
# condição necessária para algoritmos baseados em distância, como o K-Means.

variaveis_cluster = [
    "NU_NOTA_CN",
    "NU_NOTA_CH",
    "NU_NOTA_LC",
    "NU_NOTA_MT",
    "NU_NOTA_REDACAO"
]

scaler = StandardScaler()

X_scaled = scaler.fit_transform(df_valid[variaveis_cluster])

# Convertendo para DataFrame para facilitar inspeção e interpretação
X_scaled = pd.DataFrame(X_scaled, columns=variaveis_cluster)

X_scaled.head()
```

As notas das diferentes áreas do conhecimento apresentam escalas semelhantes, porém com distribuições e dispersões distintas.

## Aplicação de Algoritmos


### Clusterização com K-Means
Nesta etapa, aplica-se o algoritmo K-Means, um método particionado que busca dividir os dados em k grupos, minimizando a variabilidade interna (inércia) e maximizando a separação entre clusters.

Para determinar o número ideal de clusters, serão utilizadas métricas internas de avaliação:
- Silhouette Score
- Calinski-Harabasz
- Davies-Bouldin
- Método do Cotovelo (Inércia)

Inicialmente, será realizada a etapa de avaliação do número ideal de clusters (k).

Para isso, o algoritmo será executado para diferentes valores de k, e serão analisadas métricas internas de validação que medem coesão e separação dos grupos formados.

Essa etapa tem como objetivo identificar o valor de k que melhor representa a estrutura dos dados, antes da aplicação definitiva do modelo sobre a base completa.

```python
# Amostra para definição do número ideal de clusters
X_sample = X_scaled.sample(n=20000, random_state=42)

print("Shape da amostra utilizada para avaliação de k:", X_sample.shape)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

range_k = range(2, 9)

sil_scores = []
cal_scores = []
db_scores = []
inertias = []

for k in range_k:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X_sample)

    sil_scores.append(silhouette_score(X_sample, labels))
    cal_scores.append(calinski_harabasz_score(X_sample, labels))
    db_scores.append(davies_bouldin_score(X_sample, labels))
    inertias.append(km.inertia_)

print("Silhouette:", sil_scores)
print("Calinski-Harabasz:", cal_scores)
print("Davies-Bouldin:", db_scores)
print("Inércia:", inertias)
```

### 📌 Escolha do Número Ideal de Clusters

A análise das métricas internas de validação indica que o valor k = 2 apresenta o melhor desempenho global:

- Maior Silhouette Score (0,356), indicando melhor separação entre os grupos.
- Maior índice de Calinski-Harabasz.
- Menor índice de Davies-Bouldin, indicando maior coesão interna e melhor separação externa.

Embora a inércia continue diminuindo com o aumento de k, não se observa ponto de inflexão significativo após k = 2.

Dessa forma, optou-se por utilizar k = 2 como número ideal de clusters para a aplicação definitiva do algoritmo K-Means na base completa.

```python
from sklearn.cluster import KMeans

# Treinando K-Means final com k=2 na base completa
k_final = 2
km_final = KMeans(n_clusters=k_final, random_state=42, n_init=10)

clusters_km = km_final.fit_predict(X_scaled)

# Salvando o cluster no dataframe principal (df_valid)
df_valid["CLUSTER_KMEANS"] = clusters_km

print("Distribuição dos clusters (K-Means):")
display(df_valid["CLUSTER_KMEANS"].value_counts().sort_index())
```

### 📌 Aplicação Final do K-Means (k = 2)

O modelo K-Means foi aplicado à base completa utilizando k = 2, conforme definido na etapa de validação.

A distribuição dos participantes entre os clusters foi:

- Cluster 0: 278.160 participantes
- Cluster 1: 230.936 participantes

Observa-se uma divisão relativamente equilibrada entre os grupos, indicando que o algoritmo identificou dois perfis predominantes de desempenho na base analisada.

A próxima etapa consiste em interpretar as características de cada cluster, analisando médias de desempenho e fatores socioeconômicos associados.

```python
# Médias das notas por cluster
df_valid.groupby("CLUSTER_KMEANS")[[
    "NU_NOTA_CN",
    "NU_NOTA_CH",
    "NU_NOTA_LC",
    "NU_NOTA_MT",
    "NU_NOTA_REDACAO",
    "MEDIA_GERAL"
]].mean()
```

### 📌 Estabilidade do K-Means

Após a definição do número ideal de clusters (k = 2), realizou-se uma análise de estabilidade do algoritmo K-Means.

Como o K-Means depende de inicializações aleatórias dos centróides, diferentes sementes podem gerar agrupamentos distintos.

Para verificar a consistência da solução encontrada, o algoritmo foi executado com diferentes valores de semente aleatória, comparando-se os resultados por meio do índice Adjusted Rand Index (ARI), que mede a similaridade entre duas partições.

```python
from sklearn.metrics import adjusted_rand_score

seeds = [0, 1, 2, 42, 100]
labels_dict = {}

# Rodando K-Means com diferentes sementes
for seed in seeds:
    km = KMeans(n_clusters=2, random_state=seed, n_init=10)
    labels = km.fit_predict(X_sample)  # usamos a amostra para acelerar
    labels_dict[seed] = labels

# Comparando estabilidade (ARI entre todas as combinações)
ari_scores = []

for i in range(len(seeds)):
    for j in range(i+1, len(seeds)):
        ari = adjusted_rand_score(
            labels_dict[seeds[i]],
            labels_dict[seeds[j]]
        )
        ari_scores.append(ari)

ari_scores
```

### 📌 Análise de Estabilidade do K-Means

Os valores do Adjusted Rand Index (ARI) entre as diferentes execuções do K-Means variaram entre 0,989 e 1,0.

Como o ARI assume valores entre 0 (sem similaridade) e 1 (agrupamentos idênticos), os resultados indicam altíssima estabilidade do modelo.

Isso significa que, independentemente da semente utilizada para inicialização dos centróides, o algoritmo converge para praticamente a mesma estrutura de clusters.

Dessa forma, conclui-se que a solução com k = 2 é robusta e consistente, não sendo resultado de aleatoriedade na inicialização.

### 📌 Clusterização Hierárquica (Método Ward)

Nesta etapa, aplica-se o algoritmo hierárquico aglomerativo utilizando o critério de Ward, que busca minimizar a variância interna dos grupos durante o processo de fusão.

O objetivo é comparar os resultados com o K-Means e verificar se a estrutura de dois clusters se mantém consistente.

```python
from sklearn.cluster import AgglomerativeClustering

agg = AgglomerativeClustering(n_clusters=2, linkage='ward')
clusters_hier = agg.fit_predict(X_sample)

print("Distribuição dos clusters (Hierárquico - Ward):")
pd.Series(clusters_hier).value_counts().sort_index()
```

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

sil_hier = silhouette_score(X_sample, clusters_hier)
cal_hier = calinski_harabasz_score(X_sample, clusters_hier)
db_hier = davies_bouldin_score(X_sample, clusters_hier)

print("Silhouette (Hierárquico):", sil_hier)
print("Calinski-Harabasz (Hierárquico):", cal_hier)
print("Davies-Bouldin (Hierárquico):", db_hier)
```

### 📌 Comparação entre K-Means e Clusterização Hierárquica

A aplicação do método hierárquico (Ward) apresentou resultados consistentes com a estrutura identificada pelo K-Means.

Comparando as métricas:

- O K-Means apresentou maior Silhouette Score.
- O K-Means obteve maior índice de Calinski-Harabasz.
- O K-Means apresentou menor índice de Davies-Bouldin.

Embora o método hierárquico tenha apresentado desempenho ligeiramente inferior, os resultados foram próximos, indicando que a estrutura de dois clusters é robusta e consistente entre diferentes técnicas de agrupamento.

Dessa forma, optou-se por manter o K-Means como modelo principal devido à sua maior eficiência computacional e melhores métricas de validação.
